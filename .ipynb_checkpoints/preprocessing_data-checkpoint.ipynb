```python
import scanpy as sc
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
import anndata as ad
import pooch
```

## Reading data from the folders
## Here we have two samples IDOX (treatment) and DMSO 

```python


data_paths = {
    "idox": "Idox_out/outs/filtered_feature_bc_matrix/",
    "dmso": "DMSO_out/outs/filtered_feature_bc_matrix",
}
adatas = {}

for sample_id, data_path in data_paths.items():
    path = "/shared/space2/molina/husainm/DNA_pertu/" + data_path  # Update the path to your directory
    sample_adata = sc.read_10x_mtx(path)
    adatas[sample_id] = sample_adata

adata = ad.concat(adatas, label="sample")
```

```python
sc.pp.calculate_qc_metrics(
    adata,inplace=True, log1p=True
)
```

```python
sc.pl.violin(
    adata,
    keys=["n_genes_by_counts", "total_counts"],
    jitter=0.4,
    multi_panel=True,
    groupby="sample"
)
```

```python
sc.pl.scatter(adata, "total_counts", "n_genes_by_counts", color="sample")
```

## Normalization of data
## Common approach we use is count depth scaling with subsequent log plus one (log1p) transformation.

```python

sc.pp.normalize_total(adata, target_sum=1e4)
```

## Taking log of the data


```python
sc.pp.log1p(adata)
```

## Feature selection
## In this step we will select the high variable features(genes) by doing dimensionality reduction

```python

sc.pp.highly_variable_genes(adata, n_top_genes=2000, batch_key="sample")
```

```python
sc.pl.highly_variable_genes(adata)
```

## PCA 
## The PCA helps to reduce the dimensionality of the data 

```python

sc.tl.pca(adata)
```

## This helps to find how much PCs we are adding which can be informative,we can also see are we adding noise? you can reduce the noice by limiting number of PCs

```python
sc.pl.pca_variance_ratio(adata, n_pcs=30,)
```
